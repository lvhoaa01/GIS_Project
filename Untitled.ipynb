```python
import time
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np
import heapq
import math

# ==============================================================================
# 1. CHUẨN BỊ "SÂN ĐUA" (Tạo dữ liệu lớn giả lập)
# ==============================================================================
print("--- ĐANG KHỞI TẠO DỮ LIỆU LỚN (STRESS TEST) ---")

# Tạo mạng lưới lớn (50x50 = 2500 giao lộ) thay vì 25 điểm
GRID_SIZE = 50
G_bench = nx.grid_2d_graph(GRID_SIZE, GRID_SIZE)
for u, v in G_bench.edges():
    G_bench[u][v]['weight'] = 1.0 # Trọng số đơn giản

# Tạo 1000 điểm dữ liệu ngẫu nhiên
NUM_POINTS = 1000
all_nodes = list(G_bench.nodes())
data_bench = random.sample(all_nodes, NUM_POINTS)

print(f"-> Mạng lưới: {len(all_nodes)} nút.")
print(f"-> Dữ liệu cần phân cụm: {NUM_POINTS} điểm.")

# ==============================================================================
# 2. ĐỊNH NGHĨA HAI ĐỐI THỦ (Copy lại hàm cốt lõi ngắn gọn)
# ==============================================================================

# --- ĐỐI THỦ A: BẢN GỐC (ORIGINAL) ---
def run_original_benchmark(G, data, eps, min_pts):
    data_set = set(data)
    # 1. Tìm hàng xóm (Dijkstra trên G đầy đủ)
    neighbors_cache = {}
    for p in data:
        dist = {p: 0}; q = [(0, p)]; nbrs = []
        while q:
            d, u = heapq.heappop(q)
            if d > eps: continue
            if u in data_set: nbrs.append(u)
            for v in G.neighbors(u):
                nd = d + G[u][v]['weight']
                if nd <= eps and nd < dist.get(v, float('inf')):
                    dist[v] = nd; heapq.heappush(q, (nd, v))
        neighbors_cache[p] = nbrs
    
    # 2. Gom cụm (Có check Noise)
    labels = {}; c_id = 0
    ordered = sorted(data, key=lambda x: len(neighbors_cache[x]), reverse=True)
    for p in ordered:
        if p in labels: continue
        if len(neighbors_cache[p]) < min_pts:
            labels[p] = -1 # Noise
        else:
            c_id += 1; labels[p] = c_id
            seeds = list(neighbors_cache[p])
            while seeds:
                q = seeds.pop(0)
                if q in labels:
                    if labels[q] == -1: labels[q] = c_id
                else:
                    labels[q] = c_id
                    if len(neighbors_cache[q]) >= min_pts: seeds.extend(neighbors_cache[q])
    return labels

# --- ĐỐI THỦ B: BẢN CẢI TIẾN (OPTIMIZED) ---
def run_optimized_benchmark(G, data, eps, min_pts):
    data_set = set(data)
    # 1. Cắt tỉa đồ thị (Pruning)
    valid_edges = [(u,v) for u,v,d in G.edges(data=True) if d['weight'] <= eps]
    G_pruned = nx.Graph(); G_pruned.add_edges_from(valid_edges)
    for u,v in valid_edges: G_pruned[u][v]['weight'] = G[u][v]['weight'] # Copy weight
    
    # 2. Lọc Heuristic & Tính mật độ
    neighbors_cache = {}
    threshold = math.log(len(data))
    ordered = []
    
    for p in data:
        # Dijkstra trên G_pruned
        dist = {p: 0}; q = [(0, p)]; nbrs = []
        while q:
            d, u = heapq.heappop(q)
            if d > eps: continue # Cải tiến điều kiện dừng
            if u in data_set: nbrs.append(u)
            # Chỉ duyệt trên G_pruned (ít cạnh hơn)
            if u in G_pruned: 
                for v in G_pruned.neighbors(u):
                    nd = d + G_pruned[u][v]['weight']
                    if nd <= eps and nd < dist.get(v, float('inf')):
                        dist[v] = nd; heapq.heappush(q, (nd, v))
        
        # Lọc Heuristic
        if len(nbrs) >= threshold:
            neighbors_cache[p] = nbrs
            ordered.append(p)
            
    ordered.sort(key=lambda x: len(neighbors_cache[x]), reverse=True)
    
    # 3. Gom cụm (Bỏ qua Noise)
    labels = {}; c_id = 0
    for p in ordered:
        if p in labels: continue
        if len(neighbors_cache[p]) >= min_pts:
            c_id += 1; labels[p] = c_id
            seeds = list(neighbors_cache[p])
            while seeds:
                q = seeds.pop(0)
                if q not in labels:
                    labels[q] = c_id
                    # Cần check lại nếu q bị lọc ở bước 2
                    if q in neighbors_cache:
                        if len(neighbors_cache[q]) >= min_pts: seeds.extend(neighbors_cache[q])
                    # Nếu q không có trong cache thì phải tính lại (nhưng ở đây ta bỏ qua cho nhanh demo)
    return labels

# ==============================================================================
# 3. CHẠY ĐUA (BENCHMARK)
# ==============================================================================
EPS_TEST = 2.5
MIN_PTS_TEST = 4

print("\n--- BẮT ĐẦU CHẠY ĐUA ---")

# Đo thời gian Bản Gốc
start_time = time.time()
run_original_benchmark(G_bench, data_bench, EPS_TEST, MIN_PTS_TEST)
end_time = time.time()
time_original = end_time - start_time
print(f"1. Bản Gốc (Original):   {time_original:.4f} giây")

# Đo thời gian Bản Cải tiến
start_time = time.time()
run_optimized_benchmark(G_bench, data_bench, EPS_TEST, MIN_PTS_TEST)
end_time = time.time()
time_optimized = end_time - start_time
print(f"2. Bản Cải tiến (Opt.):  {time_optimized:.4f} giây")

# ==============================================================================
# 4. VẼ BIỂU ĐỒ SO SÁNH
# ==============================================================================
improvement = ((time_original - time_optimized) / time_original) * 100

plt.figure(figsize=(10, 6))
algorithms = ['Original (Gốc)', 'Improved (Cải tiến)']
times = [time_original, time_optimized]
colors = ['gray', '#2ca02c'] # Xám và Xanh lá

bars = plt.bar(algorithms, times, color=colors, width=0.6)

# Ghi số giây lên cột
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.4f}s', ha='center', va='bottom', fontsize=12, fontweight='bold')

plt.ylabel('Thời gian thực thi (Giây)')
plt.title(f'SO SÁNH TỐC ĐỘ XỬ LÝ (N={NUM_POINTS} điểm)\nCải thiện tốc độ: {improvement:.2f}%', fontsize=14)
plt.ylim(0, max(times) * 1.2)
plt.show()
```
